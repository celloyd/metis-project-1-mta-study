```python
clean_by_date_and_time = mta.avg_per_day_of_week_and_time(clean_data)
clean_by_date_and_time.head()
```

```python
station = '23 ST'
# Pulls entries for station of interest
for_station_line_graph = clean_by_date_and_time[
    clean_by_date_and_time.index.get_level_values('STATION') == station]
```

```python
for_station_line_graph.reset_index(inplace = True)
```

```python
# Get it graphable?
graph_it = for_station_line_graph[['DAY_INT', 'TIME', 'COMBINED']]
time_dict = {0: '.0', 4: '.17', 8: '.33', 12: '.5', 16: '.67', 20: '.83'}
graph_it['daytime_code'] = None
for i in range(0, len(graph_it)):
    graph_it.iloc[i, 3] = float(str(graph_it.iloc[i, 0]) + time_dict[graph_it.iloc[i, 1]])

graph_it                                
```

```python
sns.lineplot(data = graph_it, x = 'daytime_code', y = 'COMBINED')
ticks_x = range(0, 7)
days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
plt.xticks(ticks_x, days, rotation = 45)
plt.title('23 St turnstile traffic over average week')
plt.ylabel('foot traffic per 4-hour window (average)')
plt.xlabel('')
plt.savefig('23 St avg weekly traffic by 4 h blocks.png')
# Label days of week
```
